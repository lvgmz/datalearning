```python
import pandas as pd
```

```python
movies = pd.read_csv('../imdb_top_1000.csv')
print(movies.head())
print(movies.info())
```

```python
print(f"Total movies: {len(movies)}")
```

```python
print(movies.columns)
```

```python
print(movies.head(10))
```

```python
movies['Released_Year'] = pd.to_numeric(movies['Released_Year'], errors='coerce')

recent = movies[movies['Released_Year'] >= 2010]
print(f"Movies since 2010: {len(recent)}")
#print(movies.dtypes)
```

```python
highly_rated = movies[movies['IMDB_Rating'] > 8.0]
print(highly_rated[['Series_Title', 'IMDB_Rating']])
```

```python
movies['Runtime_min'] = movies['Runtime'].str.extract('(\\d+)').astype(float)
long_movies = movies[movies['Runtime_min'] > 120]
print(f"Movies over 2hr: {len(long_movies)}")
```

```python
print(f"Average rating: {movies['IMDB_Rating'].mean():.2f}")
```

```python
best = movies.loc[movies['IMDB_Rating'].idxmax()]
print(f"Best rated: {best['Series_Title']} ({best['IMDB_Rating']})")
```

```python
print(movies['IMDB_Rating'].describe())
```

```python
movies['decade'] = (movies['Released_Year'] // 10) * 10
by_decade = movies.groupby('decade')['IMDB_Rating'].mean()
print("\nAverage rating by decade:")
print(by_decade)
```

```python
by_cert = movies.groupby('Certificate').size()
print("\nMovies by rating:")
print(by_cert)
```

```python
top10 = movies.nlargest(10, 'IMDB_Rating')[['Series_Title', 'IMDB_Rating', 'Released_Year']]
print("\nTop 10 movies:")
print(top10)
```

```python
print("\nMissing values:")
print(movies.isnull().sum())
```

```python
clean_movies = movies.dropna(subset=['IMDB_Rating'])
print(f"After dropping nulls: {len(clean_movies)} movies")
```

```python
complex_query = movies[movies['Runtime_min'] > 120].groupby('decade')['IMDB_Rating'].agg(['mean', 'count'])
print("\nLong movies by decade:")
print(complex_query)
```

```python
star_movies = movies[movies['Series_Title'].str.contains('Star', case=False, na=False)]
print(f"\nMovies with 'Star': {len(star_movies)}")
```

```python
movies['rating_category'] = movies['IMDB_Rating'].apply(
    lambda x: 'Excellent' if x >= 8.0 else 'Good' if x >= 7.0 else 'Average'
)
print(movies[['Series_Title', 'IMDB_Rating', 'rating_category']].head(10))

print("\n✅ Pandas practice complete!")
```

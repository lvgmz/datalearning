```python
import pandas as pd
```

```python
movies = pd.read_csv('../imdb_top_1000.csv')
print(movies.head())
print(movies.info())
```

```python
print(f"Total movies: {len(movies)}")
```

```python
print(movies.columns)
```

```python
print(movies.head(10))
```

```python
recent = movies[movies['Released_Year'] >= 2010]
print(f"Movies since 2010: {len(recent)}")
```

```python
highly_rated = movies[movies['IMDB_Rating'] > 8.0]
print(highly_rated[['Series_Title', 'IMDB_Rating']])
```

```python
movies['Runtime_min'] = movies['Runtime'].str.extract('(\\d+)').astype(float)
long_movies = movies[movies['Runtime_min'] > 120]
print(f"Movies over 2hr: {len(long_movies)}")
```

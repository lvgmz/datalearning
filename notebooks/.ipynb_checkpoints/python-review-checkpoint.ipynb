```python
 # 1 Lists (arrays)
numbers = [1,2,3,4,5]
print(numbers[0])
print(numbers[-1])
print(numbers[1:3])
numbers.append(6)
print(len(numbers))
```

```python
# 2 List comprehensions
squares = [x**2 for x in numbers]
evens = [x for x in numbers if x % 2 == 0]
print(squares)
print(evens)
```

```python
# 3 Dictionaries
person = {
    'name': 'Alice',
    'age': 27,
    'city': 'Boston'
}
print(person['name'])
print(person.get('age'))
person['job'] = 'Analyst'
```

```python
# 4 Loops
for num in numbers:
    print(num * 2)
for key, value in person.items():
    print(f"{key}: {value}")
```

```python
# 5 Functions
def analyze_data(data):
    return sum(data)/len(data)
avg = analyze_data(numbers)
print(f"Average: {avg}")
```

```python
# 6 Lambda functions
double = lambda x: x * 2
print(double(5))
```
